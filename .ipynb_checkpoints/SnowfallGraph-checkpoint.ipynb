```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\maxem\Downloads\2016-2026 22840 Daily.csv")

# Keep only needed columns
df = df[['DATE', 'SNOW']]

# Convert date column
df['DATE'] = pd.to_datetime(df['DATE'])

# Remove missing snowfall values
df = df.dropna(subset=['SNOW'])

# Optional: remove days with zero snowfall
snow_days = df[df['SNOW'] > 0]

print(snow_days.head())
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['DATE'], df['SNOW'])

plt.title("Daily Snowfall 2016–2026")
plt.xlabel("Date")
plt.ylabel("Snowfall (inches)")

plt.show()
```

```python
df['YEAR'] = df['DATE'].dt.year

yearly_snow = df.groupby('YEAR')['SNOW'].sum()

yearly_snow.plot(kind='bar')

plt.title("Total Snowfall Per Year (2016–2026)")
plt.xlabel("Year")
plt.ylabel("Total Snowfall (inches)")

plt.show()
```
